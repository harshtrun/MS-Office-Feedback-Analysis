```python
import pandas as pd
import numpy as np
```

```python
target=[]
dataText=[]
with open('en-training.txt','r') as text:
    for i in text:
        feed=i.strip().split()
        if(len(feed)>1):
            target.append(feed[-1])     
            del(feed[0])
            del(feed[-1])
            dataText.append(feed) 
        else:
            print(feed)
```

```python
# dataText
```

```python
stoplist=[]
with open('stopwords.txt','r') as w:
    for i in w:
        stoplist.append(i.strip())
stoplist
```

```python
filtData=[]
for w in dataText:
    list=[]
    for i in w:
        if(i.lower() not in stoplist):
            list.append(i.lower())
    filtData.append(list)
filtData
```

```python
from collections import defaultdict
freq=defaultdict(int)
for i in filtData:
    for w in i:
        freq[w]+=1
```

```python
# freq['room']
```

```python
dataSet=[]
for i in filtData:
    data=[]
    for w in i:
        if freq[w]>1:
            data.append(w)        
    dataSet.append(data)    
```

```python
# dataSet
```

```python
import gensim 
```

```python
dict_bow=gensim.corpora.Dictionary(dataSet)
bow=[dict_bow.doc2bow(i) for i in dataSet]
indexes=len(dict_bow)
dense_bow=gensim.matutils.corpus2dense(bow,num_terms=indexes).transpose()
dense_bow
```

```python
indexes
```

```python
tfidf=gensim.models.TfidfModel(bow)
records=tfidf[bow]
dense_tfidf=gensim.matutils.corpus2dense(records,num_terms=indexes).transpose()
dataset=dense_tfidf
```

```python
for i in dataset:
    print(len(i))
```

```python
tag=[]
for i in target:
    if i=='comment':
        tag.append(0)
    elif i=='complaint':
        tag.append(1)
    elif i=='request':
        tag.append(2)
    elif i=='bug':
        tag.append(3)
    elif i=='undefined':
        tag.append(4)
    elif i=='meaningless':
        tag.append(5)
    else:
        tag.append("None")
targetNew=tag   
targetNew
```

```python
from sklearn.model_selection import KFold
kf=KFold(n_splits=10,random_state=None,shuffle=True)
```

```python
from sklearn import svm
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
from sklearn.metrics import f1_score
from sklearn.naive_bayes import GaussianNB
from sklearn.naive_bayes import MultinomialNB
from sklearn.naive_bayes import BernoulliNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
accuracies=[]
psArr=[]
rsArr=[]
fsArr=[]
```

```python
for train,test in kf.split(dataset):
#     print(train,test)
    train_set=[]
    train_tag=[]
    for ind in train:
        train_set.append(dataset[ind])
        train_tag.append(targetNew[ind])
    test_set=[]
    test_tag=[]
    for ind in test:
        test_set.append(dataset[ind])
        test_tag.append(targetNew[ind])
#     print(len(train_set),len(train_tag))
#     print(len(test_set),len(test_tag))  
#     classifier=svm.SVC(gamma='scale')
#     classifier=GaussianNB()
#     classifier=MultinomialNB()
#     classifier=BernoulliNB()
#     classifier=KNeighborsClassifier(n_neighbors=6)
#     classifier=DecisionTreeClassifier(random_state=0)
    classifier=RandomForestClassifier(n_estimators=100, max_depth=2,random_state=0)
    classifier.fit(train_set,train_tag)
    pred=classifier.predict(test_set)
    incorrect=(pred!=test_tag).sum()
    
    accuracy=(len(test_set)-incorrect)/float(len(test_set))*100
    
    ps=precision_score(test_tag,pred,average='weighted')
    rs=recall_score(test_tag,pred,average='weighted')
    fs=f1_score(test_tag,pred,average='weighted')
    accuracies.append(accuracy)
    
    psArr.append(ps)
    rsArr.append(rs)
    fsArr.append(fs)
#     print(accuracies,ps,rs,fs)
#     print(accuracy)
print("Accuracy= ",max(accuracies)) 
print("Precision Score= ",max(psArr))
print("Recall Score= ",max(rsArr))
print("f1 Score= ",max(fsArr))

```
